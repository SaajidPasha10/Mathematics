#### PMF (Probability Mass Function) It describes the probability distribution for discrete random variables. Ex: Rolling a fair dice (1-6)

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
pmf = [1/6] * 6 # [1/6,1/6....]

plt.title("PMF")
plt.bar(x,pmf)
plt.xlabel("Dice Outcome")
plt.ylabel("Probabilities")
plt.show()
```

PDF (Probability Density function)
This describes the probability distribution for continuous random variables. Each probability here is a range of values not an exact value

```python
import matplotlib.pyplot as plt
from scipy.stats import expon
import numpy as np

x = np.linspace(0,10,1000)
y = expon.pdf(x,scale=2) # scale = 1/lambda
plt.plot(x,y,label="Exponential PDF")
plt.xlabel("X")
plt.ylabel("Density")
plt.legend()
plt.show()
```

#### **CDF (Cumulative Distribution Function) **
####  Probability that a random variable 'X' is less than or equal to a certain value "x" F(x) = P(X<=x)
#### Ex: What percentage of students scored less than 70%

```python
marks = np.arange(10,100,10) # 10,20,...100
pmf = np.arange(1,len(marks)+1) # [1,2,....9]
pmf = pmf / sum(pmf)
cdf = np.cumsum(pmf)

print(f"PMF {pmf}")
print(f"CDF {cdf}")

plt.figure(figsize = (12,5))
#PMF
plt.subplot(1,2,1)
plt.bar(marks,pmf,width=5,color="skyblue")
plt.title("Discrete Random Variables(PMF)")
plt.xlabel("Marks")
plt.ylabel("Probability")


#CDF
plt.subplot(1,2,2)
plt.step(marks,cdf,color="red",where="mid")
plt.title("CDF for PMF")
plt.xlabel("Marks")
plt.ylabel("Cumulative Probability")
# Optional start
cutoff = 70
mask = marks < cutoff
print(f"Mask {mask}")
plt.fill_between(marks, cdf, step="mid", alpha=0.3, color="orange")
plt.axvline(x=cutoff, color="blue", linestyle="--", label="70% cutoff")
plt.legend()
#Optional end
plt.tight_layout()
plt.show()
```

### CDF for PDF

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Normal distribution (mean=0, std=1)
x = np.linspace(-4, 4, 1000)
pdf = norm.pdf(x, 0, 1)   # PDF
cdf = norm.cdf(x, 0, 1)   # CDF

plt.figure(figsize=(12,5))

# PDF
plt.subplot(1,2,1)
plt.plot(x, pdf, color="blue", label="PDF")
plt.fill_between(x, pdf, alpha=0.3, color="blue")
plt.title("PDF (Probability Density Function)")
plt.xlabel("x")
plt.ylabel("Density")
plt.legend()

# CDF
plt.subplot(1,2,2)
plt.plot(x, cdf, color="red", label="CDF")
plt.fill_between(x, cdf, alpha=0.3, color="red")
plt.title("CDF (Cumulative Distribution Function)")
plt.xlabel("x")
plt.ylabel("Cumulative Probability")
plt.legend()

plt.tight_layout()
plt.show()
```
